# **정형 데이터 VS 비정형 데이터**

머신러닝과 딥러닝을 보통 정형 데이터와 비정형 데이터로 나눈다.

---

정형 데이터는 보통 레코드라고 말할 수 있는, 행과 열로 잘 정렬되어 있는, 그런 일반적으로 정형화된 데이터 포맷으로 CSV나 데이터베이스, 혹은 엑셀에 저장하기 쉽다.

지금까지 배운 머신러닝 알고리즘은 정형 데이터에 최적화되어있다.

비정형 데이터는 정형 데이터와 반대로 데이터베이스나 엑셀같은 규칙적인 포맷으로 표현하기 어려운 데이터로 텍스트, 오디오, 이미지, 영상 등이 있다.

비정형 데이터는 딥러닝 알고리즘에서 뛰어난 성능을 보이고 있다.

---

정형 데이터를 딥러닝 알고리즘에, 비정형 데이터를 머신러닝 알고리즘에 물론 사용할 수 있다.

하지만 비정형 데이터를 머신러닝 알고리즘에 사용하면 성능이 좋지 않다.

그래서 보통 머신러닝은 정형 데이터, 딥러닝은 비정형 데이터를 사용하는 것이 일반적이다.

정형 데이터는 일반적으로 데이터를 가공하는 전처리 혹은 특성공학이라 부르는 작업이 쉽다. 하지만 비정형 데이터는 특성공학을 적용하기 까다롭다. 

예를 들어서, 특정 이미지에서 고양이나 강아지의 모습을 찾는다면 고양이나 강아지의 특징을 미리 정의하기 어렵다.

또한, 비정형 데이터는 규칙성을 찾기 어려워서 머신러닝 방법으로는 모델을 만들기 까다롭다.

딥러닝 알고리즘이 특징을 추출하는 작업을 자동으로 해주기 때문에 딥러닝 알고리즘은 비정형 데이터에, 머신러닝 알고리즘은 정형 데이터에 적용하는 것이 일반적이다.

note. 특징을 자동으로 추출하기 때문에 딥러닝 알고리즘을 표현학습이라고도 한다.

# **앙상블 학습(Ensemble Learing)**

정형 데이터를 다루는 데 가장 뛰어난 성능을 내는 알고리즘이 앙상블 학습이다.

앙상블 학습을 간단하게 말하면 여러 개의 모델을 합쳐서 하나의 결과를 만들어내는 것이다.

모델을 하나 훈련하는 것보다 여러 개를 학습해서 여러 개의 모델의 예측을 평균내거나 다수의 예측을 따라서 더 나은 결과를 만들어내는 것이다.

이 알고리즘은 대부분 결정 트리를 기반으로 만들어져 있다.

트리는 어떠한 규제가 없다면 과대적합이 된다. 굉장히 성능이 좋아진다. 

그래서 기본적으로 성능이 좋은 트리를 기반으로 트리를 여러 개 만들어서 과대적합을 줄일 수 있으면 성능도 좋고, 일반화 했을 때 테스트 세트에서도 성능을 높일 수 있는 그런 모델을 만들 수 있다.

또한, 앙상블 학습의 특징은 랜덤이 많다. 무작위성을 줘서 일부로 트리의 성능을 낮추는 효과를 내도록 만든다.

# **랜덤 포레스트(Random Forest)**

랜덤 포레스트는 앙상블 학습의 대표 주자 중 하나로 안정적인 성능 덕분에 널리 사용되고 있다.

결정 트리를 랜덤하게 만들어서 숲을 만드는 알고리즘이다.

그리고 각 결정 트리의 예측을 사용해 최종 예측을 만든다.

분류일경우에는 각 결정 트리에서 나오는 클래스별 확률을 모두 더해서 트리의 개수로 나눈다. 그러면 각 클래스마다 평균적인 확률이 나올 것이고 가장 큰 확률을 예측으로 사용한다.

회귀일경우에는 각 결정 트리에서 나오는 예측값을 모두 더해서 트리의 개수로 나눠서 평균한다.

# **부트스트랩 샘플(Bootstrap sample)**

image.png

랜덤 포레스트는 각 트리를 훈련하기 위한 데이터로 부트스트랩 샘플을 사용한다.

부트스트랩 샘플링이란 입력한 훈련 데이터에서 랜덤하게 샘플을 훈련 세트의 크기와 동일하게 추출하여 훈련 데이터를 만든다. 

이때 한 샘플이 중복되어 추출될 수도 있다. 

즉, 데이터 세트에서 중복을 허용하여 데이터를 샘플링하는 방식이다.

예를 들어, 1,000개 가방에서 100개씩 샘플을 뽑는다면 먼저 1개를 뽑고, 뽑았던 1개를 다시 가방에 넣는다. 이런 식으로 계속해서 100개를 가방에서 뽑으면 중복된 샘플을 뽑을 수 있다.

이렇게 부트스트랩 샘플링을 사용하면 훈련 세트가 매 번 조금씩 달라질 것이고, 중복된 샘플이 있기 때문에 성능이 그렇게 좋지 않을 것이다.

image.png

결정 트리 모델은 부모와 자식 노드 사이의 불순도 차이가 가능한 크도록 노드를 분할하며 성장시킨다.

보통은 전체 특성 중에서 불순도 차이를 가장 크게 만드는 특성을 찾아야한다.

그런데 랜덤 포레스트에서는 각 노드를 분할할 때 전체 특성 중에서 전체 특성 개수의 제곱근만큼의 일부 특성을 무작위로 고른 다음 이중에서 최선의 분할을 찾는다.

예를들어 4개의 특성이 있다면 노드마다 2개를 랜덤하게 선택하여 사용한다.

모든 노드를 만들 때마다 위의 과정을 반복한다.

※ 다만 회귀모델은 전체 특성을 사용한다. 그래서 분류와는 달리 특성 선택에서 무작위성이 포함되지 않는다.※

랜덤 포레스트는 랜덤하게 선택한 샘플과 특성을 사용하기 때문에 훈련 세트에 과대적합되는 것을 막아준다. 

당연히 각 결정 트리의 성능은 낮아지지만 낮은 성능의 트리를 묶어서 일반화하면 전체적으로 높은 성능이 나온다. 

또한 검증 세트와 테스트 세트에서 안정적인 성능을 얻을 수 있다.


# **랜덤 포레스트 훈련**

기본적으로 랜덤 포레스트는 100개의 결정 트리를 위의 방식으로 훈련한다.

당연히 매개변수로 트리의 개수를 조정할 수 있다.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split

wine = pd.read_csv('https://bit.ly/wine_csv_data')

data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()

train_input, test_input, train_target, test_target = train_test_split(data, target, test_size=0.2, random_state=42)
```

와인 데이터셋을 판다스로 불러오고 훈련 세트와 테스트 세트로 나눈다.

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

랜덤 포레스트는 사이킷런의 앙상블 패키지(sklearn.ensemble) 하위에 RandomForestClassifier 클래스를 임포트해서 사용할 수 있다.

cross_validate() 함수를 사용해서 교차 검증을 수행해보았다.

기본적으로 100개의 결정 트리를 사용하므로 n_jobs 매개변수를 -1로 지정하여 모든 CPU코어를 사용하는 것이 좋다.

기본적으로 검증 세트의 점수만 반환하는데 return_train_score 매개변수를 True로 지정하면 훈련 세트에 대한 점수도 같이 반환한다.

이렇게하면 훈련 세트와 검증 세트의 점수를 비교할 수 있어서 과대적합을 파악하는데 용이하다.

---

출력된 결과를 보면 훈련세트에 다소 과대적합된 것처럼 보인다.

```python
rf.fit(train_input, train_target)
print(rf.feature_importances_)
```

이번에는 기본 매개변수가 만족스럽다고 가정한 후, 매개변수 튜닝을 하지 않고 교차검증을 하지 않고, 훈련세트로 모델을 훈련했다.

랜덤포레스트는 결정 트리의 앙상블이기 때문에 DecisionTreeClassifier가 제공하는 중요한 매개변수를 모두 제공한다.

그래서 결정 트리의 큰 장점 중 하나인 특성 중요도를 계산한다.

랜덤 포레스트의 특성 중요도는 각 결정 트리의 특성 중요도를 취합한 것이다.

---

※ 5장 1절의 특성 중요도 ※

[0.12345626   0.86862934    0.0079144]

출력된 결과를 5장 1절의 결정 트리에서 만든 특성 중요도와 비교해보면, 각각 [알코올 도수, 당도, pH]였는데 당도의 중요도가 감소하고 알코올 도수와 pH특성의 중요도가 조금 상승했다.

이렇게 된 이유는 랜덤 포레스트가 특성의 일부를 랜덤하게 선택하여 결정 트리를 훈련하기 때문이다.

그 결과 하나의 특성에 과도하게 집중하지 않고 좀 더 많은 특성이 훈련에 기여할 기회를 얻는다.

```python
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)

rf.fit(train_input, train_target)
print(rf.oob_score_)
```

랜덤 포레스트는 훈련 세트에서 중복을 허용하여 부트스트랩 샘플을 만들어 결정 트리를 훈련한다고 했다.

이때 부트스트랩 샘플에 포함되지 않고 남는 샘플이 있는데 이런 샘플을 OOB(out of bag)샘플이라고 한다.

이 남는 샘플을 사용하여 마차 검증 세트처럼 부트스트랩 샘플로 훈련한 결정 트리를 평가할 수 있다.

그러면 검증 세트를 따로 떼놓거나 교차검증을 사용하지 않아도 검증 세트의 역할을 대신할 수 있다.

이것이 랜덤 포레스트의 또 다른 특징으로 자체적으로 모델을 평가하는 점수를 얻을 수 있다.

---

이 점수를 얻으려면 RandomForestClassifier 클래스의 oob_score 매개변수를 True로 지정해야한다.

그러면 각 결정 트리의 OOB 점수를 평균하여 반환한다.

그 점수는 oob_score_속성에 있다.

OOB점수를 사용하면 교차 검증을 대신할 수 있어서 결과적으로 훈련 세트에 더 많은 샘플을 사용할 수 있다.

# **엑스트라 트리(Extra Trees)**

랜덤 포레스트와 매우 비슷하게 동작하는데 조금 다른 알고리즘이 엑스트라 트리이다.

엑스트라 트리는 랜덤 포레스트랑 굉장히 많은 부분은 공유하고 있다.

동일하게 100개의 결정 트리를 훈련하고, 결정 트리가 제공하는 대부분의 매개변수를 지원한다.

또한 분류에서는 전체 특성 개수의 제곱근만큼의 일부 특성을 랜덤하게 선택하여 노드를 분할하는데 사용한다.

회귀에서는 전체특성을 사용한다.

---

랜덤 포레스트와 엑스트라 트리의 차이점은 부트스트랩 샘플의 사용이다.

엑스트라 트리는 부트스트랩 샘플을 사용하지 않고, 결정 트리를 만들 때 전체 훈련 세트를 사용한다.

대신 노드를 분할할 때 가장 좋은 분할을 찾는 것이 아니라 무작위로 분할한다.

만약 전체 특성에서 랜덤하게 특성을 선택하고, 그 중에서 가장 좋은 불순도 차이를 나타내는 것을 찾아서 노드를 분할한다.

하나의 결정 트리에서 무작위로 분할한다면 성능이 낮아지겠지만 많은 트리를 앙상블 하기 때문에 과대적합을 막고 검증 세트의 점수를 높이는 효과가 있다.

---

보통 엑스트라 트리가 무작위성이 좀 더 크기 때문에 랜덤 포레스트보다 더 많은 결정 트리를 훈련해야 한다.

하지만 랜덤하게 노드를 분할하기 때문에 빠른 계산 속도가 장점이다.

```python
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

엑스트라 트리는 사이킷런의 앙상블 패키지(sklearn.ensemble) 하위에 ExtraTreesClassifier 클래스를 임포트해서 사용할 수 있다.

cross_validate() 함수를 사용해서 교차 검증을 수행해보았다.

랜덤 포레스트와 비슷한 결과를 얻었다.

```python
et.fit(train_input, train_target)
print(et.feature_importances_)
```

엑스트라 트리도 랜덤 포레스트와 마찬가지로 특성 중요도를 제공한다.

결과를 보면 랜덤 포레스트와 비슷하고, 결정 트리보다 당도에 의한 의존성이 작다.

# **그래디언트 부스팅(Gradient Boosting)**

앙상블 학습에 급이 있다면 그래디언트 부스팅이 가장 급이 높은 알고리즘이라고 할 수 있다.

이름에서 볼 수 있듰이 4장에서 배운 경사 하강법을 사용하여 트리를 앙상블에 추가한다.

분류에서는 로지스틱 손실 함수를 사용하고, 회귀에서는 평균 제곱 오차 함수를 사용한다.

경사하강법을 설명할 때 손실 함수를 산의로 정의하고, 산의 경사를 따라서 가장 낮을 곳을 내려오는 것이었다. 

이때 가장 낮은 곳을 찾아 내려오는 방법은 모델의 가중치와 절편을 조금씩 바꾸는 것이었다.

결국 손실 함수에 값(손실값)들은 결국 어떤 실수값이고, 이 실수값이 낮아지도록 하는 것이다.

그래디언트 부스팅은 그 실수값이 낮아지도록 결정 트리를 계속 추가하는 것이다.

근데 그 실수값(손실값)을 낮추는 트리를 추가하기때문에 회귀 트리를 추가한다.

GradientBoostingClassifier 클래스나 GradientBoostingRegressor 클래스 둘 다 사용하는 트리는 DecisionTreeRegressor를 사용한다.

---

또한 손실 함수의 낮은 곳으로 천천히 조금씩 이동해야한다고 했다.

그레디언트 부스팅도 마찬가지이다. 

그래서 그래디언트 부스팅에는 learning rate라는 학습률 매개변수로 속도를 조정한다.

또한 max_depth 매개변수를 3으로 제한을 해서 깊이가 3인 깊이가 얕은 결정트리회귀모델 100개를 사용한다.

깊이가 얕으면 성능이 좋지는 않지만 과대적합 가능성이 낮다.

하지만 계속해서 트리를 추가하여 성능을 높일 수 있다. 결국 일반적으로 높은 일반화 성능을 기대할 수 있다.

하지만 단점이 한 개씩 모델을 만들어서 손실함수의 값을 뽑은 다음에 두 번째 트리를 추가하고, 이런식으로 순서대로 트리를 추가하기 때문에 병렬로 처리하기 어려워 훈련 속도가 느리다.

그러다보니 GradientBoostingClassifier에는 n_jobs 매개변수가 없다.

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

사용 방법은 동일하게 사이킷런의 앙상블 패키지 하위의 GradientBoostingClassifier 클래스를 임포트해서 사용하면 된다.

이번에도 cross_validate() 함수를 사용해서 교차 검증을 수행했다.

거의 과대적합이 되지 않았고, 굉장히 일반화되었다.

```python
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

트리의 개수는 n_estimators 매개변수로 지정하면 된다. default는 100이다.

학습률은 learning_rate 매개변수로 지정하면 된다. default는 0.1이다.

결정 트리 개수를 500개로, 학습률을 0.2로 학습률을 증가시키고 트리의 개수를 늘려서 교차 검증을 수행했다.

---

트리의 개수를 5배나 늘렸는데 훈련 세트의 점수와 검증 세트의 점수가 그렇게 막 차이가 나지 않는다. 과대적합을 잘 억제하고 있다.

```python
gb.fit(train_input, train_target)
print(gb.feature_importances_)
```

역시 feauter_importances_ 특성 중요도를 제공하고 있다.

결과를 보면 그래디언트 부스팅이 랜덤포레스트보다 일부 특성(당도)에 집중한다.

그래디언트 부스팅에는 subsample 매개변수가 있다.

이 매개변수는 트리 훈련에 사용할 훈련 세트의 비율을 정하는 매개변수로 default는 1.0으로 전체 훈련 세트를 사용한다.

1보다 작으면 훈련 세트의 일부를 사용하는데 이는 마치 경사 하강법 단계마다 일부 샘플을 랜덤하게 선택하여 진행하는 확률적 경사 하강법이나 미니배치 경사 하강법과 비슷하다.

# **히스토그램 기반 그레디언트 부스팅(Histogram-based Gradient Boosting)**

히스토그램 기반 그레이디언트 부스팅은 그레디언트 부스팅의 속도와 성능을 더욱 개선한 것으로 정형 데이터를 다루는 머신러닝 알고리즘 중에 가장 인기가 높은 알고리즘이다.

히스토그램이란 어떤 데이터를 일정한 구간으로 쪼개서 막대 그래프를 쭉 세워놓은 것이다.

이런식으로 훈련 데이터를 256개의 구간으로 쪼개서 특성을 변환한다. 각 구간에 대해서 학습(노드 분할)을 하는 알고리즘이다.

특성값의 범위가 제한되어있기 때문에 빠르게 학습(노드 분할)할 수 있고, 과대적합도 억제하고, 높은 성능을 낼 수 있다.

히스토그램 기반 그레디언트 부스팅은 256개의 구간 중에서 한 개를 누락된 값을 위해서 사용한다. 

따라서 어떤 샘플에 대해서 누락된 값이 있으면 따로 떼어놓은 그 구간으로 할당하기때문에 누락된 특성이 있더라도 따로 전처리할 필요가 없다.

```python
# 사이킷런 1.0 버전 아래에서는 다음 라인의 주석을 해제하고 실행하세요.
# from sklearn.experimental import enable_hist_gradient_boosting
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

사이킷런의 앙상블 패키지 하위에 HistGradientBoostingClassifier을 임포트해서 사용할 수 있다.

일반적으로 기본 매개변수에서 안정적인 성능을 얻을 수 있다.

트리의 개수를 지정할 때는 n_estimators 대신에 부스팅 반복 횟수를 지정하는 max_iter를 사용한다.

---

cross_validate() 함수로 교차검증을 수행했고, 점수를 축려했다.

과대적합을 억제하면서 그레이디언트 부스팅보다 조금 더 높은 성능을 얻었다.

```python
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)
result = permutation_importance(hgb, train_input, train_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

특성 중요도를 계산하는 히스토그램 기반 그레이디언트 부스팅뿐만아니라 다른 추정기 모델에도 적용할 수 있는 새로운 기법이 있다.

그 기법은 Permutation Importance(치환 중요도)이다.

이 기법은 특성을 하나씩 랜덤하게 섞어서 모델의 성능이 변화하는지를 관찰하여 어떤 특성이 중요한지를 계산한다.

---

n_repeats 매개변수는 랜덤하게 섞을 횟수를 지정한다. default는 5이다.

permutation_importance() 함수가 반환하는 객체는 반복하여 얻은 특성 중요도(importances), 평균(importances_mean), 표준 편차(importances_std)를 담고 있다.

평균을 출력해보면 랜덤 포레스트와 비슷한 비율임을 알 수 있다.

합이 1이 안되고, 정확도의 차이이다.

두 번째 특성을 섞으면 정확도가 평균적으로 0.23만큼이나 떨어진다는 얘기로 두번째 특성이 그만큼 중요하다는 것을 보여준다.

첫 번째 특성과 세 번째 특성은 섞어도 정확도의 손실은 0.08 정도 밖에 되지 않는다.

```python
result = permutation_importance(hgb, test_input, test_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

지금까지 특성 중요도는 훈련 세트에 대해서만 특성 중요도였다.

이 기법은 훈련 세트뿐만 아니라 테스트 세트에서도 적용할 수 있다.

테스트 세트에 적용할 수 있다는 말은 모델을 실전에 투입했을 때 어떤 특성에 관심을 둘지 예상할 수 있다는 것이다.

```python
hgb.score(test_input, test_target)
```

최종적으로 모델을 평가했을 때 단일 결정 트리보다 좋은 결과를 얻었다.
