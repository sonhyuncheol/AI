# **점진적인 학습**

저번 예제에서 럭키백 안의 7개의 생선에 대한 확률을 머신러닝 모델을 만들었다.

이번 예제에서는 그 럭키백 이벤트가 인기가 많아지면서 생선을 공급하겠다는 어민이 늘어났다.

그러다보니 샘플을 골라내는 일이 어려워졌고, 추가되는 생선의 샘플도 가지고 있지 않다.

어떤 종류의 생선이 먼저 올지도, 모든 생선이 도착할 때까지 기다릴 수도 없다.

현재 당면한 문제는 훈련 데이터가 한 번에 준비되는 것이 아니라 조금씩 전달된다는 것이다.

---

이 문제를 해결 방법은 기존의 훈련 데이터에 매번 추가되는 새로운 데이터를 활용해 모델을 훈련하는 것이다.

하지만 이 방법은 시간이 지날수록 데이터가 엄청나게 늘어나는 것이다.

또 다른 방법은 새로운 데이터를 추가할 때 이전 데이터를 버림으로써 훈련 데이터 크기를 일정하게 유지하는 것이다.

하지만 이 방법은 데이터를 버릴 때 샘플 수가 적은 데이터가 많이 손실되거나 다른 데이터에 없는 중요한 데이터가 포함되어 있을 수도 있다.

위에서 언급한 방법은 이전에 훈련한 모델을 버리고 다시 새로운 모델을 훈련하는 방식이다.

---

그러면 원래 있던 모델을 유지하면서 즉, 가중치나 절편을 유지하면서 새로운 데이터에 대해서만 조금씩 더 훈련할 수 없을까?

이런 식의 훈련 방식을 점진적 학습 또는 온라인 학습이라고 부른다.

대표적인 점진적 학습 알고리즘은 확률적 경사 하강법(Stochastic Gradient Descent)이다.



# **확률적 경사 하강법(Stochastic Gradient Descent)**

확률적 경사 하강법은 머신러닝, 딥러닝 알고리즘이 아니다.

확률적 경사 하강법은 머신러닝, 딥러닝 알고리즘을 훈련하는, 최적화하는 방법이다.

확률적은 '랜덤하게'라는 말이고, 경사는 '기울기'라는 말이고, 하강법은 '내려가는 방법'라는 말이다.

다시 말해 확률적 경사 하강법은 랜덤하게 경사를 따라 내려가는 방법이다.

캡처.PNG

가장 빨리 내려가는 방법은 가장 가파른 경사를 찾는 것이다.

경사 하강법이 바로 가장 가파른 경사를 따라 원하는 지점(최적점)에 도달하는 것이 목표다.

하지만 내려갈 때 step이 너무 크면 경사를 따라 내려가지 못하고 오히려 올라갈 수도있다.

그렇기때문에 가장 가파른 길을 찾아 내려오지만 조금씩 내려오는 것이 중요하다.

이렇게 내려오는 과정이 바로 경사 하강법 모델을 훈련하는 것이다.

캡처.PNG

경사 하강법으로 내려올 때 가장 가파른 길을 찾는 방법이 무엇일까?

훈련 세트를 사용해 모델을 훈련하기 때문에 경사 하강법도 당연히 훈련 세트를 사용하여 가장 가파른 길을 찾을 것이다.

확률적 경사 하강법은 전체 샘플을 사용하지 않고, 훈련 세트에서 '랜덤하게' 하나의 샘플을 골라 가장 가파른 길을 찾는다.

---

확률적 경사 하강법은 훈련 세트에서 랜덤하게 하나의 샘플을 선택하여 훈련한다. 즉, 조금씩 가파른 경사를 따라 내려간다.

이 과정을 훈련 세트의 샘플을 모두 사용할 때까지 반복한다.

이렇게 훈련 세트의 샘플을 한 번 모두 사용하는 과정을 1에포크(epoch)라고 부른다.

1에포크 이후에도 최적점에 도달하지 못했다면 훈련 세트에 모든 샘플을 다시 채워 넣고, 그다음 2에포크, 다시 랜덤하게 하나의 샘플을 선택해 이어서 경사를 내려간다.

이렇게 최적점에 도착할 때까지 계속 내려가면 된다.

일반적으로 경사 하강법은 훈련 세트를 한 번 이상 사용한다. 에포크를 여러 번 수행한다.

---

일반적으로 2의 배수로 개수를 정해준다.

---

전체 데이터를 사용하기 때문에 가장 안정적인 방법이 될 수 있다.

하지만 전체 데이터를 사용하면 그만큼 컴퓨터 자원을 많이 사용하게 된다.

어떤 경우는 데이터가 너무 많아서 한 번에 전체 데이터를 모두 읽을 수 없을지도 모른다.

# **로지스틱 손실 함수**

캡처.PNG

로지스틱 회귀에서 로지스틱 회귀 모델은 확률을 출력했다.

예측은 0 또는 1이지만 확률은 0 ~ 1 사이의 어떤 값도 될 수 있는, 즉 연속적이었다.

---

위의 샘플 4개의 예측 확률을 각각 0.9, 0.3, 0.2, 0.8이라고 가정했다.

첫 번째 샘플의 예측은 0.9이므로 양성 클래스의 타깃인 1과 가깝다. 즉, 오차가 작다.

0.9에 타깃인 1과 곱한 다음 음수를 붙이면 - 0.9가 되기 때문에 낮은 손실로 바꿀 수 있다.

두 번째 샘플의 예측은 0.3이므로 양성 클래스의 타깃인 1과 멀다. 즉, 오차가 크다.

0.3에 타깃인 1과 곱한 다음 음수를 붙이면 - 0.3이 되기 때문에 높은 손실로 바꿀 수 있다.

예측이 양성 클래스의 타깃인 1에 가까울수록 예측과 타깃의 곱의 음수는 점점 작아진다.

---

세 번째 샘플의 타깃은 음성 클래스 0이다.

하지만 타깃인 0을 곱하면 무조건 0이 된다.

그래서 타깃이 0인 것을 마치 양성 클래스처럼 1로 바꾼 다음 예측값도 양성 클래스에 대한 예측으로 바꾸면 된다.

즉, 세 번째 샘플의 예측은 1 - 0.2 = 0.8로, 타깃은 1로 사용하면 된다.

그래서 0.8에 타깃인 1과 곱한 다음 음수를 붙이면 - 0.8이 되기 때문에 낮은 손실로 바꿀 수 있다.

네 번째 샘플도 타깃이 0이므로 예측은 1 - 0.8 = 0.2로, 타깃은 1로 사용한다.

그래서 0.2에 타깃인 1과 곱한 다음 음수를 붙이면 - 0.2가 되기 때문에 높은 손실로 바꿀 수 있다.

---

이렇게 타깃과 가까운 값은 작은 손실로, 먼 값은 높은 손실로 만들 수 있다.

(이 방법은 로지스틱 손실 함수를 이해하기 쉽계 단계적으로 유도하는 과정에서 나온 설명)

image.png

예측 확률을 사용해 위의 방식으로 계산하면 연속적인 손실 함수를 얻을 수 있다.

여기에서 예측 확률에 로그 함수를 적용하면 더 좋다.

예측 확률의 범위는 0 ~ 1 사이인데 로그 함수는 이 사이에서 음수가 되므로 최종 손실 값은 양수가 된다.

로그 함수는 0에 가까울수록 아주 큰 음수가 되기 때문에 손실을 아주 크게 만들어 모델에 큰 영향을 끼칠 수 있다.

---

양성 클래스(타깃 = 1)일 때 손실은 -log(예측확률)로 계산한다.

확률이 0에 가까워질수록 아주 큰 양수가 되고, 1에 가까워질수록 아주 작은 양수가 된다.

음성 클래스(타깃 = 0)일 때 손실은 -log(1-예측확률)로 계산한다.

확률이 1에 가까워질수록 아주 큰 양수가 되고, 0에 가까워질수록 아주 작은 양수가 된다.


이 손실 함수를 로지스틱 손실 함수(Logistic Loss Function), 또는 이진 크로스엔트로피 손실 함수(Binary Cross-Entropy Loss Function)이라고 부른다.

다중 분류에서 사용하는 손실 함수를 크로스엔트로피 손실 함수라고 부른다.

이진 분류는 로지스틱 손실함수를, 다중 분류는 크로스엔트로피 손실 함수를 사용한다.

# **SGDClassifier**

```python
import pandas as pd

fish = pd.read_csv('https://bit.ly/fish_csv_data')
```

```python
fish_input = fish[['Weight','Length','Diagonal','Height','Width']].to_numpy()

fish_target = fish['Species'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(fish_input, fish_target, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

확률적 경사 하강법은 가장 가파른 경사를 따라 내려간다.

그렇기 때문에 가장 가파른 경사를 따라 내려가기위해서는 각각 특성마다 스케일이 같아야한다.

스케일이 다르면 경사를 공평하게 평가할 수 없다. 특성의 스케일때문에 다른 특성쪽으로 쏠릴 수 있따.

경사하강법을 사용할 때는 반드시 데이터 전처리를 통해서 표준 점수로 특성의 스케일을 변환해야한다.

항상 훈련 세트에서 학습한 통계 값으로 테스트 세트도 변환해야 한다는 것을 잊지 말아야한다.

```python
from sklearn.linear_model import SGDClassifier
```

사이킷런에서 확률적 경사 하강법을 제공하는 대표적인 분류용 클래스는 SGDClassifier이다.

당연히 회귀를 위한 확률적 경사 하강법도 있다.

사이킷런에서 확률적 경사 하강법을 제공하는 대표적인 회귀용 클래스 SGDRegressor이다.

사이킷런의 linear_model 패키지 하위에 있다.

```python
sc = SGDClassifier(loss='log', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

똑같이 객체를 만들어서 fit()메소드로 모델을 훈련하면된다.

하지만 이번에는 객체를 만들 때 2개의 매개변수를 지정해야한다.

loss는 손실 함수의 종류를 지정하는데 loss=log로 지정하면 로지스틱 손실 함수를 지정할 수 있다.

max_iter는 수행할 에포크 횟수를 지정하는데 10으로 지정하여 전체 훈련 세트를 10회 반복했다.

---

앞에서도 설명했지만 경사 하강법은 머신러닝 알고리즘이 아니다.

머신러닝 모델을 훈련하는, 최적화하는 방법 중에 하나이기때문에 SGDClassifier가 어떤 머신러닝알고리즘을 최적화할지 정해줘야하고, 그게 바로 loss함수에서 지정해줘야하는것이다.

바꿔서 말하면 어떤 손실함수를 최적화할지를 지정해줘야하는 것이 loss매개변수의 역할이다.

이렇게되면 로지스틱 손실 함수를 사용하므로 이 모델은 로지스틱회귀모델이 만든어진 것이다.

ConvergenceWarning 경고 메시지는 모델이 충분히 수렴하지 않았다는 의미이다.

이때는 max_iter 매개변수의 값을 늘려주는 것이 좋다. 오류가 아닌 경고이다.

※ 여기서 잠깐 ※

앞에서 로지스틱 손실 함수는 이진 분류에서 사용한다고 했는데 7개의 생선을 분류하는 다중 분류에서 사용했는가?

SGDClassifier는 이진 분류와 다중 분류에 모두 로지스틱 손실 함수를 사용한다.

다중 분류일 경우에는 이진 분류 모델을 여러 개를 여러 개를 연결하여 사용한다.

예를 들어 클래스가 A, B, C 3개인 경우 먼저 A를 양성, B,C를 음성 클래스로 놓고 이진 분류를 수행한다.

이런 식으로 B, C를 양성 클래스로 놓고 모델을 만들어 총 3개의 모델을 훈련한다.

예측을 할 때는 3개 중 가장 큰 값을 예측 클래스로 삼는다.

이런 방식을 OvA(one versus all)라고 한다.

```python
sc.partial_fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

확률적 경사 하강법은 점진적 학습이 가능하다.

SGDClassifier 객체를 다시 만들지 않고 훈련한 모델을 추가로 더 훈련할 수 있다.

만약 훈련한 모델에 fit()메소드를 다시 사용해서 훈련하면 이전에 학습한 가중치와 절편들을 다 버리고 새로 학습하는 것이다.

훈련한 모델에 partial_fit()메소드를 사용하면 기존에 학습한 가중치와 절편을 유지하면서 1 에포크씩 이어서 훈련한다.

아직 점수가 낮지만 에포크를 한 번 더 실행하니 정확도가 향상되었다.

이 모델을 여러 에포크에서 더 훈련하면 더 좋아질 가능성이 보인다.

하지만 얼마나 더 훈련해야할지는 확신할 수가 없어서 무작정 많이 반복할 수는 없다.

어떤 기준이 필요하다.

그리고 SGDClassifier는 미니배치 경사 하강법이나 배치 하강법을 제공하지 않는다.

# **에포크와 과대/과소적합**

확률적 경사 하강법을 사용한 모델은 에포크 횟수에 따라 과소적합이나 과대적합이 될 수 있다

에포크 횟수가 적으면 모델이 훈련 세트를 덜 학습한다.

이전에 규제에서 규제값이 커질수록 과소적합이 되고, 훈련 세트와 테스트 세트의 정확도가 모두 낮아졌따.

규제값이 작아질수록 과대적합이 되고, 훈련 세트의 정확도는 높아지고, 테스트 세트의 정확도는 낮아졌다.

그래서 중간의 최적점, 최적의 alpha값을 찾아서 모델을 다시 훈련하는 과정을 했었다.

image.png

마찬가지로 에포크도 비슷하다.

적은 에포크 횟수 동안에 훈련한 모델은 훈련 세트와 테스트 세트에 잘 맞지 않는 과소적합된 모델일 가능성이 높다.

반대로 많은 에포크 횟수 동안에 훈련한 모델은 훈련 세트에 너무 잘 맞아 테스트 세트에는 오히려 정확도가 낮은 과대적합된 모델일 가능성이 높다.

---

위의 그래프는 에포크가 진행됨에 따라 모델의 정확도를 나타낸 것이다.

훈련 세트 점수는 에포크가 진행될수록 꾸준히 증가하지만 테스트 세트 점수는 어느 순간 감소하기 시작한다.

바로 이 지점이 모델이 과대적합되기 시작하는 곳으로 과대적합이 시작하기 전에 훈련을 멈추는 것을 조기 종료라고 한다.

```python
import numpy as np

sc = SGDClassifier(loss='log', random_state=42)

train_score = []
test_score = []

classes = np.unique(train_target)
```

```python
for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target, classes=classes)
    
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
```

fit()메소드를 사용하지 않고, partial_fit()메소드만 사용해서 점진적인 학습을 했다.

하지만 partial_fit() 메소드만 사용하려면 partial_fit()메소드는 훈련 세트의 일부분만 주어질 수 있다고 가정하기 때문에 훈련 세트에 있는 전체 클래스 개수(레이블, 목록)를 전달해줘야한다.

이를 위해 np.unique() 함수로 train_target에 있는 7개 생선의 목록을 만들어서 partial_fit() 메소드의 classes 매개변수에 목록을 전달해서 훈련한다.

그리고 에포크마다 훈련 세트와 테스트 세트에 대한 점수를 기록하기 위해 2개의 리스트를 만들었다.

```python
import matplotlib.pyplot as plt

plt.plot(train_score)
plt.plot(test_score)
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

데이터가 작기 때문에 테스트 세트의 성능이 눈에 보일 정도로 떨어지지는 않고, 일정하게 유지하면서 100번째 에포크 이후부터 훈련세트와 테스트 세트의 점수가 조금씩 벌어지고 있다.

또 확실히 에포크 초기에는 과소적합되어 훈련 세트와 테스트 세트의 점수가 낮다.

이 모델의 경우 백 번째 에포크가 적절한 반복 횟수로 보이며, 100번째 에포크에서 조기 종료가 적절하다.

```python
sc = SGDClassifier(loss='log', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

SGDClassifier는 일정 에포크 동안 성능이 향상 되지 않으면 더 훈련하지 않고 자동으로 멈춘다.

tol 매개변수에서 향상될 최솟값을 지정한다

tol 매개변수를 None으로 지정하면 자동으로 멈추지 않는다.

SGDClassifier의 반복 횟수를 100에 맞추고 모델을 다시 훈련하니 훈련 세트에서 95%, 테스트에서 92%의 정확도를 보였다.

```python
sc = SGDClassifier(loss='hinge', max_iter=100, tol=None, random_state=42)
sc.fit(train_scaled, train_target)

print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

SGDClassifier의 loss 매개변수의 기본값은 'hinge'이다.
